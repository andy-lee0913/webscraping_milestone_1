```python
import pandas as pd
```

```python
df = pd.read_csv('cpi.csv')
```

```python
df
```

```python
# split the integrated column "LOCATION_INDICATOR"
df[['LOCATION','INDICATOR']] = df['LOCATION_INDICATOR'].str.split('*',expand = True)
```

```python
df
```

```python
# drop the redundant column
df = df.drop(columns = ["LOCATION_INDICATOR"], axis = 1)  
```

```python
df
```

```python
#arrange the columns
df = df[['LOCATION','INDICATOR','SUBJECT','MEASURE','FREQUENCY','TIME','Value','Flag Codes']]
```

```python
df
```

```python
# drop the blank column
df = df.drop(columns = ['Flag Codes'],axis = 1)
```

```python
df = df.dropna(how = "any")
```

```python
df
```

```python
df.to_csv("clean_cpi.csv", index = None)    #保存在当前文件夹
```
